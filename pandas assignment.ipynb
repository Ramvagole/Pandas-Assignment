```python
#Q1):-
The first function to mention is read_csv or read_excel. 
Till now I used at least one of these functions in every project. 
The functions are self-explanatory already.
They are used to read a CSV or an excel file to a pandas DataFrame format.
Here I am using the read_csv function to read the FIFA dataset:
Example:-
  df = pd.read_csv("file_name.csv")
    
It is also common to use .head() function after read_csv or read_excel to see the data frame. 
By default, it shows the first 5 rows of the DataFrame.
Example:-
 df.head()
    
When you have a big dataset like that it can be hard to see all the columns. 
using .columns function, you can print out all the columns of the dataset.
Example:-
 df.columns
    
You can drop some unnecessary columns using df.drop(). 
Example:-
 df = df.drop([columns name],axis=0 or 1)
    
This function takes as a parameter the rows and column indices and gives you the subset of the DataFrame accordingly. 
Here I am taking the first 10 rows of data and index 5th to index 10th columns.
Example:-
  df.iloc[:10, 5:10]
```

```python
#Q10):-
import pandas as pd
date=pd.date_range(start="2023-05-11" ,end="2023-05-17")
df=pd.DataFrame({"Date":date,"Sales":[100,120,130,90,105,140,88]})
df["Moving_Average"]=df["Sales"].rolling(window=7).mean()
df
```
